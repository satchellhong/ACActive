```python
def random_baseline(total_hits_in_library: int, batch_size, library_size: int = 100000,
                    total_screening_budget: int = 2000, start_size: int = 64) -> list:
    hits_found_in_theory = [0]
    n_mols_screened = 0
    while n_mols_screened < total_screening_budget:
        n = start_size if n_mols_screened == 0 else batch_size
        if n + n_mols_screened > total_screening_budget:
            n = total_screening_budget - n_mols_screened

        if n_mols_screened == 0:
            hits = 1 + (n-1) * total_hits_in_library / library_size
        else:
            hits = n * total_hits_in_library / library_size
        hits_found_in_theory.append(hits + hits_found_in_theory[-1])

        n_mols_screened += n
        total_hits_in_library -= hits
        library_size -= n

    return hits_found_in_theory[1:]
tot_dataset_hits = {'ALDH1': 4986, 'PKM2': 223, 'VDR': 239, 'IDH1': 11, 'ADRB2': 5, 'OPRK1': 9, 'KAT2A': 54, 'FEN1': 100, 'GBA': 55}
baselines = {d: {s: {k: random_baseline(tot_dataset_hits[d], k, start_size=s) for k in [64]} for s in [64]} for d in ['ALDH1', 'PKM2', 'VDR']}
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

dataset = 'PKM2'

baseline = pd.read_csv('../results/43588_2024_697_MOESM4_ESM.csv')
result = pd.read_csv('../results/results3_PKM.csv')

# baseline[(baseline['acquisition_method'] == 'Exploit (best predictions)') & (baseline['architecture'] == 'mlp') & (baseline['dataset'] == 'PKM2') \
#          & (baseline['screen_cycle'] == 14)]['enrichment'].mean()
# plt.fill_between(x, np.subtract(mean_values, std_values), np.add(mean_values, std_values), alpha=0.2)
y = [x*64 for x in range(16)]
y[15] = 1000

x_mlp = []
x_mlp2 = []
x_rf = []
x_rf2 = []
x_base = []
x_base2 = []
x_result = []
x_result2 = []
for i in range(16):
    data = baseline[(baseline['acquisition_method'] == 'Exploit (best predictions)') & (baseline['architecture'] == 'mlp') & (baseline['dataset'] == dataset) \
         & (baseline['screen_cycle'] == i)]
    ef = data['enrichment'].mean() / data['hits_discovered'].mean()
    x_mlp.append(data['enrichment'].mean())
    x_mlp2.append(data['enrichment'])
    data = baseline[(baseline['acquisition_method'] == 'Exploit (best predictions)') & (baseline['architecture'] == 'gcn') & (baseline['dataset'] == dataset) \
         & (baseline['screen_cycle'] == i)]
    ef = data['enrichment'].mean() / data['hits_discovered'].mean()
    x_base.append(data['enrichment'].mean())
    x_base2.append(data['enrichment'])
    data = baseline[(baseline['acquisition_method'] == 'Exploit (best predictions)') & (baseline['architecture'] == 'rf') & (baseline['dataset'] == dataset) \
         & (baseline['screen_cycle'] == i)]
    ef = data['enrichment'].mean() / data['hits_discovered'].mean()
    x_rf.append(data['enrichment'].mean())
    x_rf2.append(data['enrichment'])
    data = result[(result['acquisition_method'] == 'exploitation') & (result['architecture'] == 'gcn') & (result['dataset'] == dataset) \
         & (result['screen_cycle'] == i)]
    x_result.append(data['hits_discovered'].mean() * ef)
    x_result2.append(data['hits_discovered'] * ef)

x_base = np.array(x_base)
x_base2 = np.array(x_base2)
x_rf = np.array(x_rf)
x_rf2 = np.array(x_rf2)
x_mlp = np.array(x_mlp)
x_mlp2 = np.array(x_mlp2)
x_result = np.array(x_result)
x_result2 = np.array(x_result2)
plt.title(dataset)
plt.plot(y, x_base)
plt.plot(y, x_mlp)
plt.plot(y, x_rf)
plt.plot(y, x_result)
plt.fill_between(y, np.subtract(x_base2.mean(axis=1), x_base2.std(axis=1)/5), np.add(x_base2.mean(axis=1), x_base2.std(axis=1)/5), alpha=0.1)
plt.fill_between(y, np.subtract(x_rf2.mean(axis=1), x_rf2.std(axis=1)/5), np.add(x_rf2.mean(axis=1), x_rf2.std(axis=1)/5), alpha=0.1)
plt.fill_between(y, np.subtract(x_mlp2.mean(axis=1), x_mlp2.std(axis=1)/5), np.add(x_mlp2.mean(axis=1), x_mlp2.std(axis=1)/5), alpha=0.1)
plt.fill_between(y, np.subtract(x_result2.mean(axis=1), x_result2.std(axis=1)/5), np.add(x_result2.mean(axis=1), x_result2.std(axis=1)/5), alpha=0.1)
plt.ylabel('enrichment')
plt.xlabel('number of molecules')
plt.legend(['gcn', 'mlp', 'rf', 'acanet'])

plt.show()
```
